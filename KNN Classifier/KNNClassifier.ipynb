## 11. Split the train dataset into training set(75%)&validation set(25%)

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, x_test, Y_train, y_test = train_test_split(X, Y, test_size=0.25)
```

```python
X_train.shape
```

```python
x_test.shape
```

## 12. Fit KNN classifies to training data


```python
from sklearn.neighbors import KNeighborsClassifier
```

## 13. Default parameters n-neighbors-5,metric='minkowski' p=2

```python
KNN = KNeighborsClassifier(n_neighbors=5, metric ="euclidean")
```

```python
KNN.fit(X_train, Y_train)
```

## 14. Predict the validation set results

```python
Y_pred = KNN.predict(x_test)
```

```python
from sklearn.metrics import classification_report, accuracy_score 
```

## 16. Printing classification report and accuracy PYTHON

```python
print(classification_report(Y_pred, y_test))
print(f"accuracy score of knn classifier {accuracy_score(Y_pred, y_test)*100} persent")
```

```python
from sklearn.metrics import confusion_matrix
```
