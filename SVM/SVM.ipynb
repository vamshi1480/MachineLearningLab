```python
X = df.iloc[:,:-1]
```

```python
X
```

```python
Y = df["Species"]
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
Y = encoder.fit_transform(Y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=1)
```

```python
from sklearn.svm import SVC
```

```python
model = SVC(C=1)
model.fit(X_train, y_train)
```

```python
#help(SVC)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score , classification_report , confusion_matrix
```

```python
print(accuracy_score(y_pred, y_test))
```

```python
print(f"confusion matrix \n {confusion_matrix(y_pred, y_test)}")
```

```python
print(f"classification report  \n {classification_report(y_pred, y_test)}")
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV
```

```python
model = SVC()
n_folds = 5
parameters = {"C": [0.1, 1, 10, 100, 100], "kernel": ["poly", "rbf", "linear"]}
#Build the model
modelCV = GridSearchCV(model, parameters, cv=n_folds, scoring="accuracy")
modelCV.fit(X_train, y_train)
```

```python

# Best parameters and accuracy
print("Best Parameters:", modelCV.best_params_)
print("Best Cross-validation Accuracy:", modelCV.best_score_)
```

```python
from sklearn.metrics import classification_report

y_pred = modelCV.predict(X_test)
print("\nTest Set Performance:\n")
print(classification_report(y_test, y_pred, target_names=encoder.classes_))
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Extract results from GridSearchCV
results = pd.DataFrame(modelCV.cv_results_)

# Just keeping relevant columns
plot_data = results[["C", "kernel", "mean_test_score"]]

# Pivot table for heatmap (C on x-axis, kernel on y-axis)
pivot = plot_data.pivot(index="kernel", columns="param_C", values="mean_test_score")

# Plot heatmap
plt.figure(figsize=(8, 5))
sns.heatmap(pivot, annot=True, cmap="YlGnBu", fmt=".3f")
plt.title("Accuracy Heatmap for C vs Kernel")
plt.xlabel("C Value")
plt.ylabel("Kernel")
plt.show()
```
